# DoWhy refutation

Renders the refutation table from `artifacts/dowhy_refutation.json`.

```python
import json

import pandas as pd

from upliftbench.config import DOWHY_REFUTATION_JSON

result = json.loads(DOWHY_REFUTATION_JSON.read_text())
print(f"Point estimate (ATE on visit): {result['estimate']:+.6f}")
print(f"N (sampled): {result['n']:,}\n")
rows = [{"refuter": k, **v} for k, v in result["refutations"].items()]
pd.DataFrame(rows)
```
